```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-np.pi, np.pi, 256, endpoint=True)
y1 = np.sin(x)
x_min = np.min(x)
plt.plot(x, y1, label="y=sin(x)", color='b')
plt.text(x_min,
         np.sin(x_min),
         "y1=sin(x)",
         fontdict={'fontsize': 10},
         color='b')
y2 = np.cos(x)
plt.plot(x, y2, label="y=cos(x)", color='r')
plt.text(x_min,
         np.cos(x_min),
         "y2=cos(x)",
         fontdict={'fontsize': 10},
         color='r')
plt.plot(x, np.ones(np.size(x)) / 2, '--', color='y', label="=>0.5")
x1 = np.linspace(-2.5, -0.5, 100)
plt.fill_between(x1, np.sin(x1), np.cos(x1), color='purple')
plt.fill_between(x, y1, where=(np.abs(x) < 0.5) | (y1 > 0.5), color='green')
plt.legend()
plt.xticks(np.linspace(-np.pi, np.pi, 5, endpoint=True),
           ["-pi", "-pi/2", "0", "pi/2", "pi"])
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from pandas import read_excel

data = read_excel("country.xls")
name = data.columns.values[1:5]
grade = ["Gold", "Silver", "Copper"]
colors = ["Gold", "Silver", "Darkgoldenrod"]

#
plt.pie(data.sum(axis=0).values[1:], autopct='%1.1f%%', labels=name)
plt.show()

#
temp = data.loc[0:2, name].values
for i in range(3):
    plt.plot(name, temp[i], 'o--', label=grade[i])
    for x, y in zip(name, temp[i]):
        plt.text(x, y, '%.0f' % y, fontdict={'fontsize': 10})
plt.legend()
plt.show()

#
plt.subplot(211)
X = np.arange(len(name))
data.loc[0:2, name[0]].values
for i in range(3):
    plt.bar(X + 0.25 * i, temp[i], label=grade[i], 
            width=0.25, color=colors[i])
    for x, y in zip(X + 0.25 * i, temp[i]):
        plt.text(x, y, '%.0f' % y, fontdict={'fontsize': 10})
plt.legend()
plt.xticks(X + 0.25, name)

#
for i in range(3):
    x = 234 + i
    plt.subplot(x)
    plt.bar(name, temp[i], label=grade[i], color=colors[i])
    plt.xticks(rotation=90)
    for x, y in zip(X, temp[i]):
        plt.text(x, y, '%.0f' % y, fontdict={'fontsize': 8})
    plt.title(grade[i], fontdict={'fontsize': 10}, loc='center')
plt.tight_layout()
plt.show()
```
