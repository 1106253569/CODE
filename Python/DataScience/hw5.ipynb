```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#导入数据
dataset = pd.read_excel("company2.xls")

data = dataset['age'].values
plt.subplot(221)
plt.hist(data)
plt.subplot(222)
sns.distplot(data, hist=False, color='red')
sns.rugplot(data)
plt.subplot(223)
sns.kdeplot(data, shade=True, color='green')
plt.subplot(224)
sns.distplot(data, color='purple')
plt.tight_layout()
plt.show()
```

```python
sns.countplot(x='company',
              data=dataset,
              linewidth=6,
              edgecolor=sns.color_palette("dark"))
```

```python
plt.subplot(121)
sns.countplot(x='sex', data=dataset)
plt.subplot(122)
sns.countplot(y='sex', data=dataset)
plt.tight_layout()
plt.show()
```

```python
sns.jointplot(x=dataset['age'], y=dataset['company'])
plt.show()
sns.violinplot(x=dataset['age'], y=dataset['company'])
plt.show()
```

```python
def age_group(x):
    if x <= 18:
        return 'Child'
    elif 19 <= x <= 35:
        return 'Young'
    elif 36 <= x <= 59:
        return 'Middle'
    elif 60 <= x:
        return 'Old'


dataset['age_level'] = dataset['age'].map(age_group)
sns.countplot(x='age_level', data=dataset)
```

```python
sns.set(style='whitegrid')
sns.regplot(x='time', y='salary', data=dataset)
```

```python
c = dataset.corr()
print("相关矩阵如下:")
print(c)
sns.pairplot(dataset)
plt.show()
sns.heatmap(c)
plt.show()
```
